# Occupation

### Introduction:


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
users.head(10)
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
    
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
u = users.groupby('occupation').gender.sum()

for i in range(len(u)):
    msum = 0
    for j in range(len(u[i])):
        if u[i][j] == "M":
            msum += 1
    u[i] = str(msum/len(u[i]))

u.sort_values(ascending = False)    
    
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
occst = list(set(users.occupation))
data = []
for item in occst:
    data.append({'occupation': item, 'min age': (users.loc[(users['occupation'] == item)])['age'].min(), 'max age': (users.loc[(users['occupation'] == item)])['age'].max()})
result = pd.DataFrame(data)
result
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['gender', 'occupation']).mean()[['age']]
```

### Step 8.  For each occupation present the percentage of women and men

```python
u = users.groupby('occupation').gender.sum()
d = []

for i in range(len(u)):
    msum = 0
    fsum = 0
    for j in range(len(u[i])):
        if u[i][j] == "M":
            msum += 1
        else:
            fsum += 1
    d.append({'occupation': u.index[i], 'ratio Male': msum/len(u[i])*100 , 'ratio Female': fsum/len(u[i])*100})


pd.DataFrame(d)
```
